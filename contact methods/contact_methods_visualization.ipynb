```python
import pandas as pd
import numpy
import seaborn as sns
```

```python
df = pd.DataFrame({'Method of Contacting':['Walks', 'Text', 'Phone'], 'Percentage of Success (Got a Response From Right Person)':[0.16643159379407615, 0.0410150486869283, 0.027536025110572123]})
sns.barplot(x = df['Method of Contacting'], y = df['Percentage of Success (Got a Response From Right Person)'], palette='mako').set(title= 'Percentage of Success per Contact Method')
```

```python
df = pd.DataFrame({'Method of Contacting':['Walks', 'Text', 'Phone'], 'Cost of Getting One Person To Vote ($)':[168.8095238095238, 11.731153846153846, 2.694864864864865]})
sns.barplot(x=df['Method of Contacting'], y=df['Cost of Getting One Person To Vote ($)'], palette='mako').set(title= 'Cost of Getting One Person To Vote per Contact Method')
```
